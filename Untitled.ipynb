    Данные представляют собой набор характеристик, связанных с заболеваниями сердца.
    age: Возраст пациента (вещественный).
    sex: Пол пациента (0 = женщина, 1 = мужчина, бинарный).
    cp: Тип боли в груди (categorical):
        1: тип стенокардия (typical angina)
        2: атипичная стенокардия (atypical angina)
        3: неангинозная боль (non-anginal pain)
        4: бессимптомное (asymptomatic)
    trestbps: Артериальное давление в состоянии покоя (вещественный).
    chol: Уровень холестерина в крови (вещественный).
    fbs: Уровень сахара в крови натощак (0 = <= 120 мг/дл, 1 = > 120 мг/дл, бинарный).
    restecg: Результаты электрокардиографии в состоянии покоя (categorical):
        0: нормальные результаты
        1: наличие аномалий волны ST-T
        2: вероятное или определенное наличие гипертрофии левого желудочка
    thalach: Максимальная достигнутая частота сердечных сокращений (вещественный).
    exang: Появление стенокардии во время физической нагрузки (0 = нет, 1 = да, бинарный).
    oldpeak: Депрессия ST, вызванная физической нагрузкой по сравнению со значением в состоянии покоя (вещественный).
    slope: Наклон пикового сегмента ST во время физической нагрузки (categorical):
        1: наклон вверх
        2: плоский наклон
        3: наклон вниз
    ca: Количество крупных сосудов (0-3), окрашенных флуороскопией (целочисленный).
    thal: Результаты теста на стресс (categorical):
        3: нормальный
        6: исправленный дефект
        7: обратимый дефект
    target: Наличие заболевания сердца (0 = отсутствует, 1 = присутствует, бинарный).
